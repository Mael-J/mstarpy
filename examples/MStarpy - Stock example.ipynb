Stock Analysis

In this example, we will explore all the key methods used to perform an in-depth analysis of a stock.

```python
import mstarpy as ms
```

Initialize Stock to start you analysis

```python
stock = ms.Stock("FR0000121014")
```

Find the stock name

```python
stock.name
```

Find the stock isin

```python
stock.isin
```

https://www.morningstar.com/stocks/xnas/nvda/financials
